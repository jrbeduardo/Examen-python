### 4. ObtÃ©n una lista de todos los usuarios mencionados en todos los tweets

```python
df.real_text[31]
```

```python
import re
def obtener_usuarios_mencionados(text):
    return re.findall('\@\w+',text)      
```

```python
obtener_usuarios_mencionados(df.real_text[31])
```

```python
usuarios_mencionados_por_tweets = pd.DataFrame(df.real_text.apply(lambda x : obtener_usuarios_mencionados(x)))
usuarios_mencionados_por_tweets
```

```python
# construimos una lista
result = []
[result.extend(lista) for lista in df.real_text.apply(lambda x : obtener_usuarios_mencionados(x)).tolist()
]
result
```

```python
# quitamos los duplicados y contamos cuantas veces se mencionan
mencionados = {}
for x in result:
    mencionados[x] = result.count(x)
mencionados
```

```python
df_mencionados = pd.DataFrame(
    [
        [key, mencionados[key]] for key in mencionados.keys()
    ], 
    columns=['usuarios', 'menciones']
)
df_mencionados.head()
```

```python
df_mencionados.set_index('usuarios').plot(figsize=(20,8), kind='bar')
```

```python
#lista de mencionados
lista_mencionados = list(mencionados.keys())
lista_mencionados
```

### 5. ObtÃ©n el total de tweets con tendencia 0, 1 y 2 respectivamente

```python
# Creamos un diccionario para guardar los tweets  de cada tendencia 
tendencias={}
for i in range(3):
    tendencias[f'Tend {i}'] = df.tendencia[df.tendencia==i].count()
tendencias
```

```python
df_tendencias = pd.DataFrame(
    [
        [key, tendencias[key]] for key in tendencias.keys()
    ], 
    columns=['Tendencia', 'Total']
)
df_tendencias.head()
```

```python
df_tendencias.plot(figsize=(8,6), kind='bar')
```
